modify from original hymenoscyphus script (cigar_sample_sorter.py) to create flexible pipeline for different alignment and key characters

```python
def get_sam_pos(samfile_name, pos_list):
    """get input base positions from sam file"""
    geno_dict = {}
    samfile = open(samfile_name, 'r')
    for line in samfile:
        if line.startswith('@'):
            continue
        else:
            line = line.strip('\n')
            line_parts = line.split('\t')
            seq_id = line_parts[0]
            cig = line_parts[5]
            if cig.startswith('*'): # in case there are reads with no match
                continue
            else:
                seq = line_parts[9]
                cig_list = cigar_ref(cig)
                for item in pos_list:
                    # here have to add exception / if to deal with shorter sequences 
                    if len(seq) < item:
                        geno_dict.setdefault(seq_id, []).append('-') # adds gap if outside range of seq
                    else:
                        adjust = pus_finder(cig_list, item)
                        # put ? in when out of range
                        # try exception
                        try:
                            base = seq[adjust]
                        except Exception:
                            base = '-'
                            
                        
                            
                        geno_dict.setdefault(seq_id, []).append(base)
    
    return geno_dict
```

### Matching to string with regular expressions

```python
# taken from get_synapomorphies
positions = [27, 35, 68, 70, 71, 77, 81, 97, 103, 110, 111, 114, 118, 119, 120, 121, 122, 127, 133, 136, 139]
```

```python
# import genotypes from get_syns
clade_strings = {}
gtype_file = open('chenopod_trnl_genotypes.txt')
for line in gtype_file:
    line = line.strip('\n')
    parts = line.split('\t')
    clade = parts[0]
    genotype = parts[1]
    clade_strings[clade]=genotype
gtype_file.close()
```

```python
clade_strings
```

```python
import re
```

```python
amb = {'K': ['G', 'T'], 'M': ['A', 'C'], 'N': ['A', 'C', 'G', 'T'], 'R': ['A', 'G'], 'S': ['C', 'G'], 'W': ['A', 'T'], 'Y': ['C', 'T']}
```

```python
# note: use amb from below
clade_str_regex = {}
for k,v in clade_strings.items():
    newstring = ''
    for letter in v:
        if letter in amb:
            regs = amb[letter]
            newstring += '['
            for r in regs:
                newstring += r
            newstring += ']'
        else:
            newstring += letter
    clade_str_regex[k]=newstring
```

```python
clade_str_regex
```

```python
clade_regex_comp = {}
for k,v in clade_str_regex.items():
    vcomp = re.compile(r'%s' % (v))
    clade_regex_comp[k]=vcomp
```
